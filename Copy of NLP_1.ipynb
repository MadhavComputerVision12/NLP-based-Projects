# Tokenization

```python
import nltk #we said we will use this in notes
import nltk.corpus #the part of text used for nlp is called corpus
```

```python
from nltk.tokenize import word_tokenize
```

```python
sentence="By incorporating these pedagogical approaches, we create an enriching learning environment that promotes creativity, active engagement, and continuous growth. Let's embrace these approaches within the National Curriculum Framework for Foundational Stage NCERT to foster effective teaching and learning."
```

```python
nltk.download('punkt')
```

```python
tokens=word_tokenize(sentence)
print(tokens)
```

```python
print(type(tokens))
print(len(tokens))
```

```python
from nltk.probability import FreqDist
fdist=FreqDist()
```

```python
for i in tokens:
  fdist[i]=fdist[i]+1
print(fdist)
```

```python
top_10=fdist.most_common(10)
```

```python
print(top_10)
```

# Named Entity Recognition

```python
from nltk import ne_chunk
john="John lives in New York"
john_token=word_tokenize(john)
john_tags=nltk.pos_tag(john_token)
john_tags=nltk.pos_tag(john_token)
print(john_tags)
```

```python
nltk.download('maxent_ne_chunker')
nltk.download('words')
john_ner=ne_chunk(john_tags)
print(john_ner)
```

```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
doc=nlp('This is sparta!!')
```

```python
for token in doc:
  print(token.i," : ",token.text," : ",token.pos_)
```

```python
span=doc[2:5]
print(span)
```

```python
for ent in doc.ents:
  print(ent.text,ent.label_)
```

```python
print(doc.ents)
```

```python
doc=nlp("Barack Obama the former minister of USA will be vacating white house today")
```

```python
for ent in doc.ents:
  print(ent.text," : ",ent.label_)
```

```python
from spacy.matcher import Matcher
```

```python
pattern=[{'LEMMA':'vacate'},{'ORTH':'white'}]
matcher=Matcher(nlp.vocab)
matcher.add('white_pattern',[pattern])
matches=matcher(doc)
```

```python
for match_id,start,end in matches:
  matched_span=doc[start:end]
  print(matched_span.text)
```

```python
doc=nlp("2018 fifa world cup : France won!!")
```

```python
for ent in doc.ents:
  print(ent.text,ent.label_)
```

```python
pattern=[{'IS_DIGIT':True},{'LOWER':'fifa'},{'LOWER':'world'},{'LOWER':'cup'}]
matcher=Matcher(nlp.vocab)
matcher.add('new_pattern',[pattern])
matches=matcher(doc)
```

```python
for match_id,start,end in matches:
  matched_span=doc[start:end]
  print(matched_span.text)
```
